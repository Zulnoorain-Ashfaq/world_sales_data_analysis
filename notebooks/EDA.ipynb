```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('fast')
```

```python
data = pd.read_csv('../data/salesforcourse-4fe2kehu.csv', index_col=0)
data
```

```python
data['Profit'] = (data['Unit Price'] - data['Unit Cost']) * data['Quantity']
data
```

```python
data['Year'] = data['Year'].convert_dtypes('int')
data
```

```python
data.info()
```

```python
data.drop('Column1', axis=1, inplace=True)
```

```python
data.describe()
```

```python
data.Date = pd.to_datetime(data.Date)
data
```

```python
data.describe(include='object')
```

```python
revenue_per_date_sales = data.groupby('Date').sum()['Revenue']
revenue_per_date_sales.index = pd.date_range('2015-01-01',freq='D',periods=revenue_per_date_sales.shape[0])
revenue_per_date_sales
```

```python
plt.figure(figsize=(15,5))
plt.title('Total Revenue Generated Globally')
plt.plot(revenue_per_date_sales)
plt.xticks(pd.date_range('2015-01-01',freq='4M',end='2016-07-29'))
plt.axhline(revenue_per_date_sales.mean(),linestyle="--", c='red',label='mean',alpha=0.5)
plt.legend()
plt.show()
```

```python
profit_per_date_sales = data.groupby('Date').sum()['Profit']
revenue_per_date_sales.index = pd.date_range('2015-01-01',freq='D',periods=revenue_per_date_sales.shape[0])
profit_per_date_sales
```

```python
plt.figure(figsize=(15,5))
plt.title('Total Profit Generated Globally')
plt.plot(profit_per_date_sales)
plt.xticks(pd.date_range('2015-01-01',freq='4M',end='2016-07-29'))
plt.axhline(profit_per_date_sales.mean(),linestyle="--", c='red',label='mean',alpha=0.5)
plt.legend()
plt.show()
```

```python
total_revenue_generated_by_county = data.groupby('Country').sum()['Revenue']
total_revenue_generated_by_county
```

```python
def generate_bar_plot(data, title, formatter=lambda x,y: f'${x/1000000:.1f}M', sorted = None):
    if sorted:
        data = data.sort_values(ascending=True)
    fig, ax = plt.subplots()
    ax.barh(data.index,data, color='red')
    ax.xaxis.set_major_formatter(formatter)
    plt.title(title)
    plt.show()
```

```python
generate_bar_plot(total_revenue_generated_by_county,'Total Revenue Generated Per Country')
```

```python
plt.figure(figsize=(8,8))
plt.pie(total_revenue_generated_by_county,labels=total_revenue_generated_by_county.index,autopct='%d%%',startangle=60,counterclock=False)
plt.title('World Revenue',fontdict={'fontsize':15,'fontweight':'bold'})
plt.figtext(.1,.1,f'Total Revenue = ${total_revenue_generated_by_county.sum()/1000000:.2f}M',fontdict={'fontsize':13,'fontweight':'bold'})
plt.show()
```

```python
total_profit_generated_by_county = data.groupby('Country').sum()['Profit']

generate_bar_plot(total_profit_generated_by_county, 'Total Profit Generated Per Country')
plt.figure(figsize=(8, 8))
plt.pie(total_profit_generated_by_county, labels=total_profit_generated_by_county.index, autopct='%d%%',
        startangle=60, counterclock=False)
plt.title('World Profit', fontdict={'fontsize': 15, 'fontweight': 'bold'})
plt.figtext(.1, .1, f'Total Profit = ${total_profit_generated_by_county.sum() / 1000000:.2f}M',
            fontdict={'fontsize': 13, 'fontweight': 'bold'})
plt.show()
```

```python
us_data = data[data.Country == 'United States']
total_revenue_generated_by_states_in_US = us_data.groupby("State").sum()['Revenue']
total_profit_generated_by_states_in_US = us_data.groupby("State").sum()['Profit']
total_revenue_generated_by_states_in_US
```

```python
fig, ax = plt.subplots()
data_temp = total_revenue_generated_by_states_in_US.sort_values(ascending=True)
ax.barh(data_temp.index,data_temp, color='red', log=True)
# ax.xaxis.set_major_formatter(lambda x,y: f'${x/1000000:.1f}M')
plt.title('Revenue Per USA States')
plt.show()
```

```python
fig, ax = plt.subplots()
data_temp = total_profit_generated_by_states_in_US.sort_values(ascending=True)
color = ['red' if x>0 else 'blue' for x in data_temp]
ax.barh(data_temp.index,abs(data_temp), color=color, log=True)
# ax.xaxis.set_major_formatter(lambda x,y: f'${x/1000000:.1f}M')
plt.title('Profits Per USA States')
plt.legend(['negative'])
plt.show()
```
